```python
import sqlite3
```

```python
conn = sqlite3.connect("timber.db")
cursor = conn.cursor()

# Create table
cursor.execute("""
CREATE TABLE IF NOT EXISTS timber_prices (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    type TEXT NOT NULL,
    unit_price REAL NOT NULL
)
""")
```

```python
timber_data = [
    ("Teak", 1200.0),
    ("Mahogany", 950.0),
    ("Oak", 800.0),
    ("Pine", 300.0),
    ("Cedar", 450.0),
    ("Spruce", 350.0),
    ("Birch", 500.0),
    ("Walnut", 1100.0),
    ("Rosewood", 1500.0),
    ("Bamboo", 250.0)
]


cursor.executemany("""
INSERT INTO timber_prices (type, unit_price)
VALUES (?, ?)
""", timber_data)

conn.commit()
```

```python
cursor.execute("SELECT * FROM timber_prices")
rows = cursor.fetchall()

for row in rows:
    print(row)
```

```python
conn.close()
```
